```python
# file_names = glob.glob('sample_data/*.csv')
# save_file_path = 'Updated_sample_data/'

file_names = glob.glob('/home/cloudcraftz/Github/Sample_NIFTY_EIS_DATA/2021/*.csv')
save_file_path = '/home/cloudcraftz/Github/NIFTY_EIS_DATA_Updated/2021/'
```

```python
def initial_preprocess(file_names, save_file_path):
    
    for file_name in file_names:

        data = pd.read_csv(file_name)
        columns = ['Date Time','UnixTimefrom 1-1-1980','ExchToken','BidPrice','BidQty','AskPrice','AskQty',
                   'TTq','LTP','TotalTradedPrice','Instrument','ExpiryDate','ExpiryTime','Strike','Type']
        data = data.shift(1)
        first_row_values = data.columns.to_list()
        # need to handle case : when column values are same like 0.0 again 0.0 it will make it as 0.0.1
        corrected_first_row_values = []
        for v in first_row_values:
            if v.count('.') == 2:
                corrected_first_row_values.append(v[:-2])
            else:
                corrected_first_row_values.append(v)
        data.loc[0] = corrected_first_row_values
        data.columns = columns
        data.to_csv(save_file_path+file_name.split('/')[-1], index = False)
```

```python
from datetime import datetime
from modules._utils import date_difference

dates = expiry_less_than_five.index
c = 0

thursday_corrput_files_with_len_more_than_three = []
thursday_corrput_files_with_len_less_than_four = []

not_thursday_corrput_files_with_len_more_than_one = []
not_thursday_corrput_files_with_len_less_than_two = []

thrusday_correct_files = []
not_thrusday_correct_files = []

exception_files = []
# If today is Thursday (1 = Mon, 2 = Tue, 3 = Wen ...)
for idx, val in tqdm(enumerate(expiry_less_than_five['filename'].values)):
    
    try:
        
        file_name = val.split('/')[-1]
        date = file_name.split('_')[1]
        date_format = date[:4] + '-' + date[4:6]+'-' +date[6:]
        
        if expiry_less_than_five.index[idx].isoweekday() == 4:
            # if thursday, we should have nearest_weekly, nearest_monthly, second_weekly , second_monthly
            expiry_dates_list = [pd.to_datetime(expiry) for expiry in expiry_less_than_five['ExpiryDate'].loc[dates[c]]]
            
            if len(expiry_dates_list) > 3:
                # 
                nearest_weekly_eis_data = HistoricalData(source='eis_data', name = 'EIS DATA', underlying_instrument='BANKNIFTY',start_date=date, end_date=date, expiry_type='nearest_weekly')
                nearest_weekly_eis_data.load_market_data()
                nearest_weekly = eis_data.getData()

                nearest_monthly_eis_data = HistoricalData(source='eis_data', name = 'EIS DATA', underlying_instrument='BANKNIFTY',start_date=date, end_date=date, expiry_type='nearest_monthly')
                nearest_monthly_eis_data.load_market_data()
                nearest_monthly = eis_data.getData()

                second_weekly_eis_data = HistoricalData(source='eis_data', name = 'EIS DATA', underlying_instrument='BANKNIFTY',start_date=date, end_date=date, expiry_type='second_weekly')
                second_weekly_eis_data.load_market_data()
                second_weekly = eis_data.getData()

                second_monthly_eis_data = HistoricalData(source='eis_data', name = 'EIS DATA', underlying_instrument='BANKNIFTY',start_date=date, end_date=date, expiry_type='second_monthly')
                second_monthly_eis_data.load_market_data()
                second_monthly = eis_data.getData()

                if len(nearest_weekly) == 0 or len(nearest_monthly) == 0 or len(second_weekly) == 0 or len(second_monthly) == 0:
                    thursday_corrput_files_with_len_more_than_three.append(val)
                else:
                    thrusday_correct_files.append(val)

            else:
                #save corrupt file
                thursday_corrput_files_with_len_less_than_four.append(val)

            c += 1

        else:
            # if not thursday , we should have nearest weekly and nearest monthly
            expiry_dates_list = [pd.to_datetime(expiry) for expiry in expiry_less_than_five['ExpiryDate'].loc[dates[c]]]
            if len(expiry_dates_list) > 1:
                # need to check dates expiry dates are nearest_weekly, nearest_monthly

                nearest_weekly_eis_data = HistoricalData(source='eis_data', name = 'EIS DATA', underlying_instrument='BANKNIFTY',start_date=date, end_date=date, expiry_type='nearest_weekly')
                nearest_weekly_eis_data.load_market_data()
                nearest_weekly = eis_data.getData()

                nearest_monthly_eis_data = HistoricalData(source='eis_data', name = 'EIS DATA', underlying_instrument='BANKNIFTY',start_date=date, end_date=date, expiry_type='nearest_monthly')
                nearest_monthly_eis_data.load_market_data()
                nearest_monthly = eis_data.getData()

                if len(nearest_weekly) == 0 or len(nearest_monthly) == 0:

                    not_thursday_corrput_files_with_len_more_than_one.append(val)
                else:
                    not_thrusday_correct_files.append(val)

            else:
                # #save corrupt file
                not_thursday_corrput_files_with_len_less_than_two.append(val)
            c += 1
    except Exception as e:
        print(e)
        print(val)
        exception_files.append(val)
```
